```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load my data: can be accuracy or pss
# Add the headers: S short, M median, L long, E early, O ontime, Late, comp 1,2,3,4,5,6 
conditions = ['SE1','ME1','LE1','SO1','MO1','LO1','SL1','ML1','LL1',
                         'SE2','ME2','LE2','SO2','MO2','LO2','SL2','ML2','LL2',
                         'SE3','ME3','LE3','SO3','MO3','LO3','SL3','ML3','LL3',
                         'SE4','ME4','LE4','SO4','MO4','LO4','SL4','ML4','LL4',
                         'SE5','ME5','LE5','SO5','MO5','LO5','SL5','ML5','LL5',
                         'SE6','ME6','LE6','SO6','MO6','LO6','SL6','ML6','LL6']
acc = pd.read_csv('/Users/t.z.cheng/Google_Drive/Research/Delaydoesmatter/real_exp/exp2/data_analysis/acc.csv',
                  names=conditions)
acc.head(48)
```

```python
mean = np.mean(acc)
stderr = np.std(acc)/(len(acc)**0.5)
```

```python
fig, ax = plt.subplots()
delay = ('Short','Median','Long')
y_pos = np.arange(len(conditions))
plt.bar(y_pos,mean, width=0.5, yerr=stderr, color='orange',align='center')
plt.xticks(y_pos, conditions)
plt.ylabel('Accuracy')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.title('Accuracy against Delay Length')
```

```python
## Onset time: early, ontime, late
early_trials = acc[['SE1','ME1','LE1','SE2','ME2','LE2','SE3','ME3','LE3','SE4','ME4','LE4','SE5','ME5','LE5','SE6','ME6','LE6']]
ontime_trials = acc[['SO1','MO1','LO1','SO2','MO2','LO2','SO3','MO3','LO3','SO4','MO4','LO4','SO5','MO5','LO5','SO6','MO6','LO6']]
late_trials = acc[['SL1','ML1','LL1','SL2','ML2','LL2','SL3','ML3','LL3','SL4','ML4','LL4','SL5','ML5','LL5','SL6','ML6','LE6']]
onset_mean = [np.mean(np.mean(early_trials)),np.mean(np.mean(ontime_trials)),np.mean(np.mean(late_trials))]
n = (len(acc)**0.5)
onset_stderr = [np.std(np.mean(early_trials,axis=1))/n,np.std(np.mean(ontime_trials,axis=1))/n,np.std(np.mean(late_trials,axis=1))/n]

fig, ax = plt.subplots()
onset = ('Early', 'Ontime', 'Late')
y_pos = np.arange(3)
plt.bar(y_pos,onset_mean, width=0.5, yerr=onset_stderr, color='orange',align='center')
plt.xticks(y_pos, onset)
plt.ylabel('Accuracy')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.title('Accuracy against Delay Length')
```

```python
## Delay length: short, median, long
short_trials = acc[['SE1','SE2','SE3','SE4','SE5','SE6','SO1','SO2','SO3','SO4','SO5','SO6','SL1','SL2','SL3','SL4','SL5','SL6']]
median_trials = acc[['ME1','ME2','ME3','ME4','ME5','ME6','MO1','MO2','MO3','MO4','MO5','MO6','ML1','ML2','ML3','ML4','ML5','ML6']]
long_trials = acc[['LE1','LE2','LE3','LE4','LE5','LE6','LO1','LO2','LO3','LO4','LO5','LO6','LL1','LL2','LL3','LL4','LL5','LL6']]
delay_mean = [np.mean(np.mean(short_trials)),np.mean(np.mean(median_trials)),np.mean(np.mean(long_trials))]
n = (len(acc)**0.5)
delay_stderr = [np.std(np.mean(short_trials,axis=1))/n,np.std(np.mean(median_trials,axis=1))/n,np.std(np.mean(long_trials,axis=1))/n]


fig, ax = plt.subplots()
delay = ('Short','Median','Long')
y_pos = np.arange(3)
plt.bar(y_pos,delay_mean, width=0.5, yerr=delay_stderr, color='orange',align='center')
plt.xticks(y_pos, delay)
plt.ylabel('Accuracy')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.title('Accuracy against Delay Length')
```

```python
## Delay length: short, median, long x Onset time: early, ontime, late
early_short = acc[['SE1','SE2','SE3','SE4','SE5','SE6']]
early_median = acc[['ME1','ME2','ME3','ME4','ME5','ME6']]
early_long = acc[['LE1','LE2','LE3','LE4','LE5','LE6']]

ontime_short = acc[['SO1','SO2','SO3','SO4','SO5','SO6']]
ontime_median = acc[['MO1','MO2','MO3','MO4','MO5','MO6']]
ontime_long = acc[['LO1','LO2','LO3','LO4','LO5','LO6']]

late_short = acc[['SL1','SL2','SL3','SL4','SL5','SL6']]
late_median = acc[['ML1','ML2','ML3','ML4','ML5','ML6']]
late_long = acc[['LL1','LL2','LL3','LL4','LL5','LL6']]

means_early_delay = [np.mean(np.mean(early_short)),np.mean(np.mean(early_median)),np.mean(np.mean(early_long))]
means_ontime_delay = [np.mean(np.mean(ontime_short)),np.mean(np.mean(ontime_median)),np.mean(np.mean(ontime_long))]
means_late_delay = [np.mean(np.mean(late_short)),np.mean(np.mean(late_median)),np.mean(np.mean(late_long))]

n_groups = 3
fig, ax = plt.subplots()
y_pos = np.arange(n_groups)
bar_width = 0.15
opacity = 0.8

rects1 = plt.bar(y_pos-bar_width,means_early_delay, bar_width, yerr=delay_stderr, color='orange',label = 'Early')
rects2 = plt.bar(y_pos,means_ontime_delay, bar_width, yerr=onset_stderr, color='red',label = 'Ontime')
rects3 = plt.bar(y_pos+bar_width,means_late_delay, bar_width, yerr=onset_stderr, color='pink',label = 'Late')

plt.xticks(y_pos, delay)
plt.ylabel('Proportion short')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.title('Proportion short against Delay Length')
plt.legend()
```
